```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

# Histograms of exponential random numbers

```python
random_exp = np.random.exponential(scale=(1/4), size=15)
# Numpy has a different parameter than R: scale = 1 / lambda
# See docs for explaination

random_exp
```

```python
_ = plt.hist(random_exp)
```

```python
for i in range(1,5):
    plt.subplot(2,2, i)
    plt.hist(np.random.exponential(scale=(1/4), size=15))
plt.show()
```

# Q-Q plot

```python
measurements = np.random.normal(size=100)
measurements
```

```python
_ = stats.probplot(measurements, dist="norm", plot=plt)
```

# Random variables from distributions against Normal Q-Q plot

## Normal

```python
x = np.random.normal(loc=3,scale=5, size=10)
_ = stats.probplot(x, dist='norm', plot=plt)
```

## Binomial

```python
x = np.random.binomial(n=50, p=0.25, size=40)
_ = stats.probplot(x, dist='norm', plot=plt)
```

## Uniform

```python
x = np.random.uniform(low=-3, high=3, size=60)
_ = stats.probplot(x, dist='norm', plot=plt)
```

## Poisson

```python
x = np.random.poisson(lam=350, size=200)
_ = stats.probplot(x, dist='norm', plot=plt)
```
